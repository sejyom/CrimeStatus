■ 구하고자 하는 문제

성별 위험 변화도는 이미 모두 백분률로 잘 정리가 되어있기 때문에,

위험이 많이 줄었다 = -2
위험이 약간 줄었다 = -1
10년 전과 비슷하다 = 0
위험이 상당히 커졌다 = 1
위험이 매우 커졌다 = 2
로 점수를 매겨

남자 / 여자 별로 응답값에 해당 점수를 모두 곱한 후 합을 구하여
어떤 성별이 더 위험을 많이 느끼는지 확인해보고,
음수인지 양수인지를 따져 10년 전보다 위험해졌는지 안전해졌는지 확인해보자.

```python
#원하는 테이블만 분리

df = data.loc['성별']
df
```

```python
# 값에 따른 점수를 계산하고,
# 새로운 테이블을 추가하여 옆에 만들기

male = df.loc['남자', '위험이 많이 줄었다'] * -2
male = male + df.loc['남자', '위험이 약간 줄었다'] * -1
male = male + df.loc['남자', '위험이 상당히 커졌다'] * 1
male = male + df.loc['남자', '위험이 매우 커졌다'] * 2

female = df.loc['여자', '위험이 많이 줄었다'] * -2
female = female + df.loc['여자', '위험이 약간 줄었다'] * -1
female = female + df.loc['여자', '위험이 상당히 커졌다'] * 1
female = female + df.loc['여자', '위험이 매우 커졌다'] * 2

# df['위험점수'] = [male, female]
# 위처럼 하면 한 번에 되는데 경고뜨길래 두번에 나눠서 하겠음!
df.loc['남자', '위험점수'] = male
df.loc['여자', '위험점수'] = female
df
```

```python
# 1. 남성 집단은 10년 전에 비해 지금이 더 위험해졌다고 생각하는가?

df.loc['남자', '위험점수'] > 0
```

```python
# 2. 여성 집단은 10년 전에 비해 지금이 더 위험해졌다고 생각하는가?

df.loc['여자', '위험점수'] > 0
```

```python
# 3. 남성 집단과 여성 집단 중 어느 집단이 10년 전에 비해 더 많이 위험해졌다고 생각하는가?

score = df.loc['남자', '위험점수']

if (score > df.loc['여자', '위험점수']) :
    print('남성이 비교적 위험하다고 느낌')
else :
    print('여성이 비교적 위험하다고 느낌')
```
